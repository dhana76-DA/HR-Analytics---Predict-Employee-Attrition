**Promotion Category vs Attrition Rate**

```python
#Categorizing Promotion Lag
def promotion_status(years):
    if years <= 2: return 'Recent (0-2 yrs)'
    if years <= 7: return 'Mid-Range (3-7 yrs)'
    return 'Stagnant (8+ yrs)'

employee_data['PromotionCategory'] = employee_data['YearsSinceLastPromotion'].apply(promotion_status)

#Promotion Category vs Attrition Rate
plt.figure(figsize=(10, 6))
attrition_rate = employee_data.groupby('PromotionCategory')['Attrition'].value_counts(normalize=True).unstack() * 100
attrition_rate.plot(kind='bar', stacked=True, color=['#4daf4a', '#e41a1c'])
plt.title('Attrition Percentage by Promotion Category')
plt.ylabel('Percentage of Employees')
plt.xticks(rotation=0)
plt.show()
```

```python
# Preprocessing
le = LabelEncoder()
for col in employee_data.select_dtypes(include=['object']).columns:
    employee_data[col] = le.fit_transform(employee_data[col])

# Modeling (Decision Tree)
X = employee_data.drop(['Attrition', 'SalaryBand'], axis=1)
y = employee_data['Attrition']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y)

model = DecisionTreeClassifier(max_depth=5)
model.fit(X_train, y_train)
predictions = model.predict(X_test)

# Evaluation
print(f"Model Accuracy: {accuracy_score(y_test, predictions):.2f}")
print(classification_report(y_test, predictions))

# Feature Importance
importances = pd.Series(model.feature_importances_, index=X.columns).sort_values(ascending=False)
importances.head(5).plot(kind='barh')
plt.title('Top 5 Drivers of Attrition')
plt.show()
```

Monthly income, overtime, total working years, Distance form home and years since last promotion are the reasons for the attrition.

**Reasoning**:
To calculate the attrition rate per department, I need to group the data by 'Department' and 'Attrition', count the values, unstack the 'Attrition' column to get 'Yes' and 'No' counts separately, and then compute the percentage of 'Yes' attrition for each department.



```python
department_attrition_rate = employee_data.groupby(['Department', 'Attrition']).size().unstack(fill_value=0)
department_attrition_rate['AttritionRate'] = (department_attrition_rate['Yes'] / (department_attrition_rate['No'] + department_attrition_rate['Yes'])) * 100
print(department_attrition_rate)
```

**Reasoning**:
I need to visualize the attrition rates for 'Research & Development' and 'Human Resources' departments using a bar plot and then determine which of these two has the highest rate. First, I will filter the `department_attrition_rate` to include only these two departments.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter for specific departments
specific_departments_attrition = department_attrition_rate.loc[['Research & Development', 'Human Resources']]

# Visualize the attrition rates
plt.figure(figsize=(8, 5))
sns.barplot(x=specific_departments_attrition.index, y=specific_departments_attrition['AttritionRate'], palette='viridis')
plt.title('Attrition Rate for Research & Development vs Human Resources')
plt.xlabel('Department')
plt.ylabel('Attrition Rate (%)')
plt.ylim(0, specific_departments_attrition['AttritionRate'].max() * 1.2)

for index, row in specific_departments_attrition.iterrows():
    plt.text(index, row['AttritionRate'] + 0.5, f"{row['AttritionRate']:.2f}%", color='black', ha="center", va='bottom')

plt.show()

# Identify department with highest attrition rate among the two
highest_attrition_dept = specific_departments_attrition['AttritionRate'].idxmax()
highest_attrition_rate = specific_departments_attrition['AttritionRate'].max()

print(f"Among 'Research & Development' and 'Human Resources', the department with the highest attrition rate is: {highest_attrition_dept} with {highest_attrition_rate:.2f}%")
```
